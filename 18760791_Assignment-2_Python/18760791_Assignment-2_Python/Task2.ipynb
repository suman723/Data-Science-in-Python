```python
# Loading libraies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
```

### Loading pre-processed dataset for each of the four stocks

```python
# Loading datasets of four stocks
appledf = pd.read_csv('apple.csv')
fbdf = pd.read_csv('facebook.csv')
alphadf = pd.read_csv('alphabet.csv')
amzdf = pd.read_csv('amazon.csv')
```

### Further pre-processing to form datetime columns for each dataset

```python
# pre-processing datasets for converting dates to datetime
appledf['Date'] = pd.to_datetime(appledf[['Day', 'Month', 'Year']].astype(str).agg('-'.join, axis=1))
appledf = appledf.sort_values(by='Date',ascending=True)
fbdf['Date'] = pd.to_datetime(fbdf[['Day', 'Month', 'Year']].astype(str).agg('-'.join, axis=1))
fbdf = fbdf.sort_values(by='Date',ascending=True)
alphadf['Date'] = pd.to_datetime(alphadf[['Day', 'Month', 'Year']].astype(str).agg('-'.join, axis=1))
alphadf = alphadf.sort_values(by='Date',ascending=True)
amzdf['Date'] = pd.to_datetime(amzdf[['Day', 'Month', 'Year']].astype(str).agg(''.join, axis=1),format='%d%b%Y',errors='coerce')
amzdf = amzdf.dropna(axis=0)
amzdf['Open'] = amzdf['Open'].astype(float)
amzdf['High'] = amzdf['High'].astype(float)
amzdf['Low'] = amzdf['Low'].astype(float)
amzdf['Close'] = amzdf['Close'].astype(float)
amzdf = amzdf.sort_values(by='Date',ascending=True)
appledf = appledf.set_index('Date')
fbdf = fbdf.set_index('Date')
alphadf = alphadf.set_index('Date')
amzdf = amzdf.set_index('Date')
```

Now, in the analysis although the stocks are compared from the given data and there trends shown, a more detailed analysis can be done by decomposing the time series of stocks into components like seasonality and trend. Decomposition can be done in two ways: Additive and Multiplicative decomposition. In Additive decompositive the stock value is represented as sum of the stated component with rabdom variation. In multiplicative method the same components are multiplied to represent the stock.
