```python
# User defined functions to load Stock data and save in html format
import requests
import pandas as pd
import numpy as np
def htmlparser(pageurl):
    url = requests.get(pageurl)
    htmltext = url.text
    return htmltext

def savehtml(htmltext,fname):
    f = open(fname+'.html', 'w')
    f.write(htmltext)
    f.close
```

### Apple Stock data collection and preparation

```python
# Parsing html webpage of Apple Inc stock
htmltext = htmlparser('http://mlg.ucd.ie/modules/COMP30760/stocks/data-aapl.html')
htmltext
```

```python
# Saving html file of Apple data
fname = 'apple'
savehtml(htmltext,fname)
```

```python
# Loading Apple stock data from html file
data = pd.read_html('apple.html')
data
```

```python
# Pre-processing time series data to combine into 1 dataframe
appledf = data[0].append(data[1]).append(data[2])
appledf
```

```python
# Missing values exploration
appledf.isna().sum()
```

```python
# filling missing values with column medians and saving to a csv format
appledf = appledf.fillna(appledf.median().to_dict())
appledf.to_csv('apple.csv',index=False)
```
